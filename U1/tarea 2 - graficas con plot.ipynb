```python
"""
Develop six engineering problems using different types of  graphs. 
"""

# 1)

"""
Graficar la posición de un objeto en caída libre a lo largo del tiempo 
utilizando la ecuación de la posición y(t) = y_0 - (1/2)*gt^2$.

Ver el cambio continuo de una variable física respecto al tiempo.
"""
import matplotlib.pyplot as plt
import numpy as np

g = 9.81
y0 = 100
t = np.linspace(0, 6, 100)
y = y0 - 0.5 * g * t**2

plt.plot(t, y, color='blue', linewidth=2)
plt.title('Caída Libre de un Objeto')
plt.xlabel('Tiempo')
plt.ylabel('Altura')
plt.grid(True)
plt.show()
```

```python
# 2)
""" 
Un laboratorio prueba la resistencia de 10 muestras de un nuevo polímero. 
Se mide el esfuerzo aplicado (MPa) y la deformación resultante (%).

Analizar la dispersión de mediciones experimentales reales o empíricas.
"""
import matplotlib.pyplot as plt

# Datos de ensayos de laboratorio
deformacion = [1.2, 2.3, 3.1, 4.5, 5.2, 6.0, 7.1, 8.3, 9.1, 10.2]
esfuerzo = [15, 28, 40, 52, 58, 70, 82, 91, 93, 105]

plt.scatter(deformacion, esfuerzo, color='red', marker='x', label='Muestras')

plt.title('Ensayo de Tensión - Esfuerzo vs Deformación')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

```python
# 3)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 600)

y = np.sin(x) * np.cos(5*x)

plt.plot(x, y, color='red', label=r'$y = \sin(x)\cos(5x)$')

plt.axhline(0, color='black', linestyle=':', linewidth=1)
plt.axvline(0, color='black', linestyle=':', linewidth=1)

plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
# 4)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 300)

y = x**3
dy = 3 * x**2

plt.plot(x, y, color='purple', label='F(y): $x^3$')
plt.plot(x, dy, '--', color='orange', label='Derivada: $3x^2$')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Análisis de una Función y su Derivada')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

```python
# 5)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)

# 2. Creamos la malla bidimensional (Meshgrid) como en tus apuntes
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')

superficie = ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Paraboloide')

plt.show()
```

```python
# 6)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 12, 400)
y = x**2

x_limite = np.linspace(0, 10, 200)
y_limite = x_limite**2

plt.plot(x, y, color='blue', label='y = $x^2$')

# Rellenar el espacio entre la curva y el eje X desde 0 hasta 10
plt.fill_between(x_limite, y_limite, color='pink', alpha=0.5)
plt.xlim(0, 12)
plt.ylim(0, 150)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
