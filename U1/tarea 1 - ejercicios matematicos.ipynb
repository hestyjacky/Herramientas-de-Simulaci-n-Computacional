```python
"""
REALIZAR 10 EJERCICIOS MATEMATICOS DE UN NIVEL COMPLEJO MEDIO
"""

# 1)
"""
un sistema de control mide la temperatura (T en C°), 
y la presion (P en atm) de un reactor químico cono una tupla (T, P)
"""
from sympy import symbols
T, P = symbols('T, P')

T = 120
P = 5.5

mediciones = (T, P)
match mediciones:
    case (T, P) if T > 100 and P > 5.0:
        print('Estado Critico')
    case (T, P) if T > 100 or P > 4.0:
        print('Estado Alerta')
    case _:
        print('Estable')
```

```python
# 2)
"""
Se simula la velocidad (v) de un objeto en caída libre
usando un paso de tiempo de 1 segundo (dt = 1). La ecuación 
de actualización en cada segundo es:

v = v + 9.8 * dt - (0.1 * v)

El bucle debe continuar ejecutándose mientras la velocidad sea
menor a 15.0 m/s y el tiempo acumulado (t) sea menor a 5 segundos.
"""
from sympy import symbols
v, t, dt= symbols('v, t, dt')

v = 0.0
t = 0
dt = 1

iteraciones = 0

# while - calcula cuántas iteraciones exactas realiza y 
# con qué valor de tiempo t termina.
while (t < 5) and (v < 15.0):
    iteraciones += 1
    v = v + (9.8 * dt) - (0.1 * v)
    t += dt
    # print(f"Segundo {t}: Velocidad = {v:.2f} m/s")

# print("-" * 30)
print(f"\tTotal de iteraciones es {iteraciones}, en {t} segundos, a la velocidad {v:.2f} m/s")
```

```python
# 3)
"""
Se lee el valor de una resistencia eléctrica en Ohmios.
Queremos clasificar el material de la siguiente manera:
"""
r = float(input('Introduzca un valor: '))

match r:
    case float() as r if (r < 0.001):
        print('El material es superconductor.')
    case float() as r if (r >= 0.001 and r <= 1.0):
        print('El material es conductor estándar.')
    case float() as r if (r >= 1.0):
        print('El material es aislante.')
        
# Escribe la estructura utilizando
# case float() as r if ... para evaluar la variable resistencia = 0.005.
```

```python
# 4)
"""
Un sensor óptico devuelve la longitud de onda de un haz de luz, 
pero a veces la envía como texto. Tenemos la variable ...
"""
longitud_onda = 532

match longitud_onda:
    # Caso 1: Si de origen fuera un número entero
    case int() as l_num if 400 <= l_num <= 700:
        print('Luz visible')
        
    # Caso 2: Si viene como texto (str), se valida y se convierte internamente
    case str() as l_str if l_str.isdigit() and int(l_str) == 532:
        print('Láser verde comercial')
        
    # Caso comodín por seguridad
    case _:
        print('Fuera del espectro o formato no reconocido')
```

```python
# 5)
"""
Un vector de fuerzas en dos dimensiones se recibe como una lista [Fx, Fy]
"""
from sympy import symbols
fx, fy = symbols('fx, fy')

[fx, fy] = [0, 50]

match [fx, fy]:
    case [0, 0]:
        print('Equilibrio estático')
    case [0, int() as fy] if fy != 0:
        print('Puramente vertical')
    case [int() as fx,0] if fx != 0:
        print('Puramente horizontal')
    case _:
        print('Bidimensional angular')
```

```python
# 6)
"""
La telemetría de un motor llega en un diccionario: 
"""
telemetria = {'RPM': 4500, 'estado': 'ESTABLE'}

match telemetria:
    case {'estado': 'CRITICO'}:
        print('Apagado de Emergencia')
    case {'RPM': int() as c} if c > 4000:
        print('Regimen de alta potencia')
```

```python
# 7)
"""
En un experimento de física nuclear, la posición de
una partícula se registra como una tupla (X, Y). 
El eje central es X=0.

Si la partícula está en X = 0 y su altura Y es mayor a 100, 
se encuentra en el "Área Norte".

Si su altura Y = 0, está en la "Línea de colisión".

Tu reto: Programa las condiciones utilizando una coordenada 
de prueba posicion = (0, 120) aplicando asignación de variables 
lógicas en los patrones.
"""
from sympy import symbols
X, Y = symbols('X, Y')

X = 0
Y = 120

posicion = (X, Y)

match posicion:
    case (0, int() as Y) if Y > 100:
        print('Area norte')
    case (int(), 0):
        print('Línea de colisión')
    case _:
        print('Desconocido')
```

```python
# 8)
from sympy import symbols, diff, solve

t = symbols('t')
s = -t**3 + 12*t**2 + 30*t

# La velocidad v(t) es la derivada de la posición respecto al tiempo
v = diff(s, t) 
# print(f"Ecuación de velocidad v(t): {v}")

# Hallamos los tiempos críticos igualando la velocidad a cero v(t) = 0
tiempos_criticos = solve(v, t)
print(f"Raíces encontradas: {tiempos_criticos}")

# Extraemos la primera raíz del tiempo crítico
t_critico = float(tiempos_criticos[1]) # Tomamos la raíz positiva

# Evaluamos si el tiempo es viable
if t_critico >= 0:
    altura_maxima = s.subs(t, t_critico)
    print(f"\n[ÉXITO] El punto más alto a los {t_critico:.2f}s")
    print(f"Altura máxima alcanzada: {float(altura_maxima):.2f}m.")
else:
    print("\n[ERROR] El modelo matemático arrojó un tiempo negativo inválido.")
```

```python
# 9)
from sympy import symbols, solve

F, m, a = symbols('F m a')

ecuacion = m * a - F

# Datos:Fuerza = 100 N, 
# Aceleración = -5 m/s²
masa_despejada = solve(ecuacion.subs({F: 100, a: -5}), m)
resultado_masa = float(masa_despejada[0])

print(f"Masa calculada: {resultado_masa}kg")

if resultado_masa > 0:
    print("Resultado válido: El objeto tiene una masa real.")
else:
    print("¡Error físico! La masa no puede ser negativa o cero.")
```

```python
# 10)
from sympy import symbols, solve

V, I, R = symbols('V I R')
# (I * R = V)
ecuacion = I * R - V

# Voltaje = 120V, 
# Resistencia = 15 Ohmios
# Despejamos la corriente (I) sustituyendo los datos conocidos
corriente_despejada = solve(ecuacion.subs({V: 120, R: 15}), I)
amperaje = float(corriente_despejada[0])

print(f"Corriente calculada: {amperaje} A")

match amperaje:
    case float() as amp if amp > 5.0:
        print("¡PELIGRO! La corriente supera los 5A. El fusible se va a fundir.")
    case float() as amp if amp > 0:
        print("Operación segura. El circuito funciona correctamente.")
    case _:
        print("Error: Circuito abierto o sin energía.")
```
